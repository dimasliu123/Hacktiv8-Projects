# Naive Bayes Bernoulli :

# Penjelasan parameter :

- Alpha : penghalusan parameter
- Binarize : nilai threshold untuk binary classification dari fitur sample.
- fit_prior : model akan belajar melalui kelas probabilitas atau tidak.
- class_prior : distribusi probabilitas.

Jika fit_prior tidak digunakan, maka class_prior yang di gunakan.

```python
nb = BernoulliNB()

param_nb = {
    'binarize' : [0, 1, 2, 3],
    'fit_prior' : [True, False],
    'alpha' : [1.0, 2.0, 3.0]
}
CV_nb = GridSearchCV(nb , param_nb, scoring='f1', cv=5)

CV_nb.fit(nbX_train, nby_train)
nb_best = CV_nb.best_estimator_
nb_pred = CV_nb.predict(nbX_val)
```

```python
print("Naive Bayes Train Score : ", nb_best.score(nbX_train, nby_train))
print("Naive Bayes Test Score : ", nb_best.score(nbX_val, nby_val))
```

# Logistic Regression :

# Penjelasan parameter :
Penalty : regularisasi


C : cost untuk menghindar misclassification (terlalu besar maka bisa overfit).


Solver : optimizer
- Newton-cg: menghitung dengan konsep kalkulus (turunan parsial ke-2)
- Lbfgs : sama seperti newton-cg, tapi perhitungan meng-exclude inverse matrix, dan di hitung dengan evaluasi gradien.

Criterion :
- Gini : menghitung nilai dari 0 -> 0.5
- Entropy : menghitung nilai dari 0 -> 1

Perbedaan terbesar Dari gini & entropy : cara mereka mengolah/menerima informasi.

```python
lr = make_pipeline(StandardScaler(), LogisticRegression(solver='liblinear'))

param_lr = {
    'logisticregression__penalty' : ['l1', 'l2'],
    'logisticregression__C' : [0.1, 1, 1.5]
}
CV_lr = GridSearchCV(lr , param_lr, scoring='f1', cv=5)

CV_lr.fit(X_train,y_train)
lr_best = CV_lr.best_estimator_
lr_pred = CV_lr.predict(X_val)
```

```python
print("Logistic Regression Train Score : ", lr_best.score(X_train, y_train))
print("Logistic Regression Test Score : ", lr_best.score(X_val, y_val))
```

# K Nearest Neighbors :

# Penjelasan parameter :
N_neighbors : jumlah dari tetangga/lingkungan.


Weights :
- uniform : daerah dari setiap model memiliki nihai yang sama.
- distance : menghitung nilai dari segi jarak.

Algorithm :
- ball_tree	: algoritma lingkaran.
- kd_tree : algoritma pohon.

N_jobs : 

jumlah/berapa nilai yang untuk mencari dataset si neighbors mencari dataset. (-1 berarti memakai semua processors).

```python
knn = KNeighborsClassifier()

param_knn = {
    'n_jobs' : [1, 2],
    'n_neighbors' : [1, 3],
    'weights' : ['distance', 'uniform'],
    'algorithm' : ['ball_tree', 'kd_tree']
}
CV_knn = GridSearchCV(knn , param_knn, scoring='f1', cv=5)

CV_knn.fit(X_train,y_train)
knn_best = CV_knn.best_estimator_
knn_pred = CV_knn.predict(X_val)
```

```python
print("KNN Train Score : ", knn_best.score(X_train, y_train))
print("KNN Test Score : ", knn_best.score(X_val, y_val))
```

# XGBoost :

# Penjelasan parameter :
- Booster : â€˜gbtreeâ€™, jenis algoritma dari XGBoost.


- base_score : Di gunakan untuk mem-balance algoritma jika suatu dataset memiliki sÄ±fat imbalance.


- Eta/learning-rate : jumlah secara numeric untuk optimisasi algoritma (biasanya dilihat dari loss function).


- n_estimators : jumlah pohon/akar dari algoritma tersebut.


- max_depth : maksimum kedalaman dari pohon tersebut.


- num_parallel_tree : Berapa pohon yang di buat untuk model tersebut.


- min_child_weight : ketentuan nilai/berat minimum yang di kirimkan ke akar selanjutnya.


- monotone_constraints : membuat sifat model tersebut menjadi datar/lurus seperti anak tangga untuk di jadikan batas XGBoost saat proses pembelajaran (Optimasi model).


- Verbose : ekspresi dari model. (0 : silent, 1 : warning, 2 : info, 3 : debug.

```python
xgboost = XGBClassifier(booster='gbtree', use_label_encoder=False, random_state = 50, base_score= 0.3, monotone_constraints = 1)

param_xgb = {
    'eta' : [0.3],
    'n_estimators' : [50],
    'max_depth' : [7],
    'min_child_weight' : [1.9]
}
CV_xgb = GridSearchCV(xgboost , param_xgb, scoring='f1', cv=5)

CV_xgb.fit(X_train,y_train)
xgb_best = CV_xgb.best_estimator_
xgb_pred = CV_xgb.predict(X_val)
```

```python
print("XGBoost Train Score : ", xgb_best.score(X_train, y_train))
print("XGBoost Test Score : ", xgb_best.score(X_val, y_val))
```

# Penjelasan parameter :
objective :
- reg:tweedie : regresi tweedie, menggunakan konsep tweedie distribution (positive skew & bersifat diskrit)

Updater :
- shotgun : Penjelasan di bawah.
- coord_descent (Coordinate descent) : optimisasi fungsi dengan turunan koordinat untuk mencari secara berurutan untuk menemukan fungsi minimum.

feature_selector :
- cyclic : Mengseleksi fitur dan melalui fitur satu per satu.
- shuffle : Mengseleksi fitur secara random sebelum setiap pembaruan.

learning-rate :
- jumlah secara numerik untuk optimisasi algoritma (biasanya dilihat dari loss function).

tweedie_variance_power :
- ketentuan untuk nilai varian dari distribusi tweedie.

```python
xgblin =XGBClassifier(booster='gblinear',objective='reg:tweedie', use_label_encoder=False, random_state = 100)

param_xgb = {
    'updater' : ['shotgun'],
    'base_score' : [0.75, 0.8, 0.9],
    'feature_selector' : ['cyclic', 'shuffle'],
    'tweedie_variance_power' : [1.0, 1.5, 1.99],
    'learning_rate' : [1.0, 1.2]
}
CV_xgblin = GridSearchCV(xgblin , param_xgb, scoring='f1', cv=5)

CV_xgblin.fit(X_train,y_train)
xgblin_best = CV_xgblin.best_estimator_
xgblin_pred = CV_xgblin.predict(X_val)
```

# Definisi/Arti dari Confusion Matrix untuk model :

- TP = Klien sudah berlangganan dan ter-prediksi sudah berlangganan.

- FN = Klien sudah berlangganan namun ter-prediksi belum berlangganan.

- FP = Klien belum berlangganan namun ter-prediksi sudah berlangganan.

- TN = Klien belum berlangganan dan ter-prediksi belum berlangganan.

Karena untuk FN & FP sama-sama merupakan hal yang vital untuk kasus tersebut, akan di fokuskan untuk melihat nilai F1-Score, dimana F1-Score merupakan kombinasi dari kalkulasi Precision & Recall.

Menilai model terbagus dari segi F1-Score, Precision & Recall dari 2 variabel (no & yes), karena dari semua model di atas yang di evaluasi, semua model memiliki nilai yang gak jauh beda dari segi akurasi.

Namun, yang membedakan ensemble learning dengan algoritma yang lain, mereka memiliki sifat tersendiri untuk mengurangi nilai bias dan varian. Ensemble learning juga memiliki kelebihan lain dimana ia mengubah weak learners menjadi strong learners di fitur tersebut.

Di samping kelebihan ensemble learning, saya juga menyadari kalau ensemble learning sangat mudah untuk overfit jika tidak di tuning dengan benar & hati-hati. Tuning di ensemble learning sangat vital untuk membuat ensemble learning menjadi model yang usable dan bagus. 

# Perbedaan Random Forest dan Algoritma Boosting
Random Forest memiliki sifat Bagging, dimana kerja/voting dilakukan dengan cara parallel. berbeda dengan model Boosting, dilakukan dengan cara sekuensial.

# Mengapa XGBoost Linear bisa mencetak nilai yang tinggi juga ?

XGBoost memiliki 1 fitur di hyperparameternya (objective=reg:tweedie), dimana menggunakan distribusi Tweedie (positive skew).

Tweedie Distribution merupakan distribusi probabilitas dimana bisa di gunakan untuk variabel yang bersifat diskrit maupun kontinu.

secara singkat, Distribusi Tweedie merupakan gabungan dari distribusi Poisson (diskrit) dan Gamma (kontinu).

updater : 'shotgun' juga membantu, dimana menggunakan konsep shotgun sequencing :
- Memecah fitur menjadi kecil lalu diurutkan secara individu, lalu model mencari pattern/urutan untuk menyusun dan memprediksi(y) secara sekuensial(konsep boosting).


Note : Shotgun sequencing sering di gunakan untuk teknik laboratorium.
