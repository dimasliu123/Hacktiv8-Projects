# Linear Regression Model :

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

# Profit :

Variable : X = Distance, y = Price

```python
X = profit.iloc[:, 2].values
y = profit.iloc[:, 1].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)

X_train= X_train.reshape(-1, 1)
X_test = X_test.reshape(-1,1)
y_train= y_train.reshape(-1, 1)

lrProfit = LinearRegression()
lrProfit.fit(X_train, y_train)

y_pred = lrProfit.predict(X_test)
```

# Train :

```python
plt.scatter(X_train, y_train, color= 'b')
plt.plot(X_train, lrProfit.predict(X_train), color='r')
```

# Test :

```python
plt.scatter(X_test, y_test, color='g')
plt.plot(X_train, lrProfit.predict(X_train), color='r')
```

```python
print('RMSE :', np.sqrt(metrics.mean_absolute_error(y_test,y_pred)))
```

```python
lyftprof = profit.loc[profit['cab_type'] == 'Lyft']

XL = lyftprof.iloc[:, 2].values
yL = lyftprof.iloc[:, 1].values

X_trainL, X_testL, y_trainL, y_testL = train_test_split(XL, yL, test_size = 1/3, random_state = 0)

X_trainL= X_trainL.reshape(-1, 1)
X_testL = X_testL.reshape(-1,1)
y_trainL= y_trainL.reshape(-1, 1)

lrProfitLyft = LinearRegression()
lrProfitLyft.fit(X_trainL, y_trainL)
```

```python
y_predL = lrProfitLyft.predict(X_testL)
```

# Linear Regression Train result on Lyft profit based on distance:

```python
plt.scatter(X_trainL, y_trainL, color= 'b')
plt.plot(X_trainL, lrProfit.predict(X_trainL), color='r')
```

# Linear Regression Test result on Lyft profit based on distance:

```python
plt.scatter(X_testL, y_testL, color= 'g')
plt.plot(X_testL, lrProfit.predict(X_testL), color='r')
```

```python
print('RMSE :', np.sqrt(metrics.mean_absolute_error(y_testL,y_predL)))
```

# Uber Linear Regression :

```python
uberprof = profit.loc[profit['cab_type'] == 'Uber']

XU = uberprof.iloc[:, 2].values
yU = uberprof.iloc[:, 1].values

X_trainU, X_testU, y_trainU, y_testU = train_test_split(XU, yU, test_size = 1/3, random_state = 0)

X_trainU= X_trainU.reshape(-1, 1)
X_testU = X_testU.reshape(-1,1)
y_trainU= y_trainU.reshape(-1, 1)

lrProfitUber = LinearRegression()
lrProfitUber.fit(X_trainU, y_trainU)
```

```python
y_predU = lrProfitUber.predict(X_testU)
```

# Linear Regression Train result on Uber price based on distance:

```python
plt.scatter(X_trainU, y_trainU, color= 'b')
plt.plot(X_trainU, lrProfitUber.predict(X_trainU), color='r')

plt.xlabel('Distance')
plt.ylabel('Price')

plt.show()
```

# Linear Regression Test result on Uber price based on distance:

```python
plt.scatter(X_testU, y_testU, color= 'g')
plt.plot(X_testU, lrProfitUber.predict(X_testU), color='r')

plt.xlabel('Distance')
plt.ylabel('Price')

plt.show()
```

```python
print('RMSE :', np.sqrt(metrics.mean_absolute_error(y_testU,y_predU)))
```

# Conclusion :

Based on the Linear Regression on the result above, we can conclude that Lyft has a slightly higher price based on their distance.

It is approved too if we check the mean between the 2 price based on their brand. 

If we divided the 2 variable (distance, price) to get the true value between the 2 category, it is true that Lyft has a slightly higher price than Uber based on their distance.
