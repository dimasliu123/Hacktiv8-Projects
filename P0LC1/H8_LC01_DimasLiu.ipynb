# Keterangan Kolom :

time(s): Detik ke- pengukuran


V_m: Voltage/tegangan yang terukur tiap waktu


Current (A): Kuat arus terukur tiap waktu


Cycle: Nomor siklus charge-discharge baterai


Cap: Kapasitas baterai tiap waktu

Ohm : hambatan listrik 

Rumus :

R = V / I

P = V * I

```python
df['Ohm(R)'] = df['V_m'] / df['Current (A)']
df['Watt (P)'] = (df['V_m'] * 2) / df['Ohm(R)']
df.head()
```

- Di kolom atas, saya membuat kolom baru untuk mencari hambatan listrik, R(Ohm) dan daya listrik, P(Watt).

- Di sini, saya hanya mau melihat nilai yang di atas 0.5

```python
hasil = [] # List kosong untuk di buat menjadi dataframe
for nilai in df['Cycle'] :
    if nilai % 2 == 0 :
        hasil.append("Genap")
    elif nilai % 2 == 1:
        hasil.append("Ganjil")
    else :
        print("Tidak ada")

df['Hasil'] = hasil
data = df
```

# Cycle yang ganjil :

```python
kapganjil = data[data['Hasil']=='Ganjil']
kapganjil
```
