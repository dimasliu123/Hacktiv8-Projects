# Modelling :
- Membuat suatu model untuk memprediksi Rating Performa orang tersebut.

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.pipeline import make_pipeline
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.feature_selection import chi2
from sklearn.metrics import plot_roc_curve
```

```python
le = LabelEncoder()
```

```python
dfs = df
```

```python
dfs['Department'] = le.fit_transform(df['Department'])
dfs['Gender'] = le.fit_transform(df['Gender'])
dfs['EducationField'] = le.fit_transform(df['EducationField'])
dfs['MaritalStatus'] = le.fit_transform(df['MaritalStatus'])
```

```python
X = dfs.iloc[:,[0,5,9,12,17,20]]
y = dfs['PerformanceRating']
y = le.fit_transform(y)
```

# Chi Square Test untuk seleksi fitur :
- HA : data bagus untuk label (untuk di jadikan model)
- H0 : data tidak bagus untuk label (untuk dijadikan model)

```python
f_stat, p_val = chi2(X,y)
p_val
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=1)
```

# Penjelasan parameter :
- eval_set : untuk mengevaluasi performa model di luar data train.
- n estimator : jumlah pohon
- max depth : kedalaman pohon
- scale pos weight : hyperparams untuk seberapa imbalance label tersebut.
- eta : learning rate

```python
eval_set = [(X_val, y_val)]
xgb = make_pipeline(StandardScaler(), XGBClassifier(random_state=42, use_label_encoder=False, 
                                                    verbosity=0, eval_set = eval_set))

                    
param_xgb = {
    'xgbclassifier__n_estimators' : [80, 90],
    'xgbclassifier__max_depth' : [5, 6],
    'xgbclassifier__scale_pos_weight' : [0.3, 0.4],
    'xgbclassifier__eta' : [0.1, 0.2]
}
CV_xgb = GridSearchCV(estimator= xgb , param_grid= param_xgb, scoring='f1', cv=5)

CV_xgb.fit(X_train, y_train)
xgb_best = CV_xgb.best_estimator_
xgb_pred = CV_xgb.predict(X_test)
```

```python
print("Best parameter : ", xgb_best)
```

```python
print(classification_report(y_test, xgb_pred))
print(confusion_matrix(y_test, xgb_pred))
```

```python
print("Cross Validation Train Set : ", cross_val_score(xgb_best, X_train, y_train, scoring='f1').mean())
print("Cross Validation Validation Set :", cross_val_score(xgb_best, X_val, y_val, scoring='f1').mean())
print("Cross Validation Test Set : ", cross_val_score(xgb_best, X_test, y_test, scoring='f1').mean())
```
