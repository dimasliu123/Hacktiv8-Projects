# Modelling :
The model will be used to predict whether an individuals will die or not based on the features.

```python
from sklearn.metrics import confusion_matrix, classification_report, plot_roc_curve
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size= 0.20, random_state = 0)
```

# Random Forest :

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(criterion='entropy', random_state = 200, bootstrap=True, n_estimators = 150, class_weight='balanced')

param_rfc = {
    'max_depth' : [4, 5, 6]
}
CV_rfc = GridSearchCV(rfc , param_rfc, scoring='precision', cv=5)

CV_rfc.fit(X_train,y_train)
rfc_best = CV_rfc.best_estimator_
rfc_pred = CV_rfc.predict(X_val)
```

```python
print("Best parameter : ", rfc_best)
print(classification_report(y_val, rfc_pred))
print(confusion_matrix(y_val, rfc_pred))
```

```python
print(cross_val_score(rfc_best, X_train, y_train, scoring='precision', cv=5).mean())
print(cross_val_score(rfc_best, X_val, y_val, scoring='precision', cv=5).mean())
```

```python
rfc.get_params().keys()
```

# Gradient Boosting :

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc = GradientBoostingClassifier(random_state = 72, n_estimators=85)

param_gbc = {
    'learning_rate' : [0.6, 0.7],
    'max_depth' : [5, 6, 7]
}
CV_gbc = GridSearchCV(gbc , param_gbc, scoring='precision', cv=5)

CV_gbc.fit(X_train,y_train)
gbc_best = CV_gbc.best_estimator_
gbc_pred = CV_gbc.predict(X_val)
```

```python
print("Best parameters : ", gbc_best)
print(classification_report(y_val, gbc_pred))
print(confusion_matrix(y_val, gbc_pred))
```

```python
print(cross_val_score(gbc_best, X_train, y_train, scoring='precision').mean())
print(cross_val_score(gbc_best, X_val, y_val, scoring='precision').mean())
```

```python
gbc.get_params().keys()
```

# AdaBoost :

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada = AdaBoostClassifier(random_state = 52, n_estimators=80)

param_ada = {
    'learning_rate' : [0.1, 0.2, 0.3]
    
}
CV_ada = GridSearchCV(ada , param_ada, scoring='precision', cv=5)

CV_ada.fit(X_train,y_train)
ada_best = CV_ada.best_estimator_
ada_pred = CV_ada.predict(X_val)
```

```python
print("Best parameters : ", ada_best)
print(classification_report(y_val, ada_pred))
print(confusion_matrix(y_val, ada_pred))
```

```python
print('AdaBoost Cross Validation (Precision) Train Score : ', cross_val_score(ada_best, X_train, y_train).mean())
print('AdaBoost Cross Validation (Precision) Test Score : ', cross_val_score(ada_best, X_val, y_val).mean())
```

```python
ada.get_params().keys()
```

# XGBoost :

```python
from xgboost import XGBClassifier
```

```python
xgboost =XGBClassifier(use_label_encoder=False, random_state= 85, gamma = 1)

param_xgb = {
    'n_estimators' : [50, 60, 70],
    'scale_pos_weight' : [0.7, 0.8],
    'max_depth' : [4, 5],
    'eta' : [0.5, 0.6]
}
CV_xgb = GridSearchCV(xgboost , param_xgb, scoring='precision', cv=5)

CV_xgb.fit(X_train,y_train)
xgb_best = CV_xgb.best_estimator_
xgb_pred = CV_xgb.predict(X_val)
```

```python
print("XGBoost Cross Validation Precision Score : ",cross_val_score(xgboost, X_train, y_train).mean())
print("XGBoost Cross Validation Precision Score : ",cross_val_score(xgboost, X_val, y_val).mean())
```

# XGBoost Early Stopping :

```python
xgb = XGBClassifier(random_state =50, eta=0.6, gamma=1, max_depth=5, n_estimators=50, scale_pos_weight=0.8)
xgb.fit(X_train, y_train, eval_set=[(X_train, y_train), (X_val, y_val)], early_stopping_rounds=20)
```

```python
xgbearlystop_pred = xgb.predict(X_val)
```

```python
xgb_auc = roc_auc_score(y_val, xgb_pred)
rfc_auc = roc_auc_score(y_val, rfc_pred)
ada_auc = roc_auc_score(y_val, ada_pred)
gbc_auc = roc_auc_score(y_val, gbc_pred)
xgbstop_auc = roc_auc_score(y_val, xgbearlystop_pred)
```

```python
xgb_fpr, xgb_tpr, _ = roc_curve(y_val, xgb_pred)
rfc_fpr, rfc_tpr, _ = roc_curve(y_val, rfc_pred)
ada_fpr, ada_tpr, _ = roc_curve(y_val, ada_pred)
gbc_fpr, gbc_tpr, _ = roc_curve(y_val, gbc_pred) 
xgbstop_fpr, xgbstop_tpr, _ = roc_curve(y_val, xgbearlystop_pred)
```

# ROC Curve :

```python
plt.figure(figsize=(7,5))

plt.plot(xgb_fpr, xgb_tpr, marker='.', label='XGBoost (AUCROC = %0.3f)' % xgb_auc)
plt.plot(rfc_fpr, rfc_tpr, marker='.', label='Random Forest (AUCROC = %0.3f)' % rfc_auc)
plt.plot(ada_fpr, ada_tpr, marker='.', label='Adaboost (AUCROC = %0.3f)' % ada_auc)
plt.plot(gbc_fpr, gbc_tpr, marker='.', label='Gradient Boosting (AUCROC = %0.3f)' % gbc_auc)
plt.plot(xgbstop_fpr, xgbstop_tpr, marker='.', label='XGBoost Early Stop (AUCROC = %0.3f)' % xgbstop_auc)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend();
```

```python
xgbprec, xgbrec, xgbthres =precision_recall_curve(y_val, xgb_pred)
rfcprec, rfcrec, rfcthres =precision_recall_curve(y_val, rfc_pred)
adaprec, adarec, adathres =precision_recall_curve(y_val, ada_pred)
gbcprec, gbcrec, gbcthres =precision_recall_curve(y_val, gbc_pred)
xgbesprec, xgbesrec, xgbesthres =precision_recall_curve(y_val, xgbearlystop_pred)
```

# Precision-Recall Curve :

```python
plt.figure(figsize=(7,5))

plt.plot(xgbprec, xgbrec, marker='.', label='XGBoost')
plt.plot(rfcprec, rfcrec, marker='.', label='Random Forest')
plt.plot(adaprec, adarec, marker='.', label='AdaBoost')
plt.plot(gbcprec, gbcrec, marker='.', label='Gradient Boosting')
plt.plot(xgbesprec, xgbesrec, marker='.', label='XGBoost Early Stop')

plt.xlabel('Recall Curve')
plt.ylabel('Precision Curve')
plt.legend();
```

Dari hasil ROC Curve, dan Precision-Recall Curve, bisa di simpulkan kalau Gradient Boosting(pertama) dan Random Forest(kedua) memiliki hasil yang terbaik.

Gradient Boosting memiliki nilai FP & FN terendah dari segi confusion matrixnya.

# Kenapa Gradient Boosting bisa memiliki nilai tertinggi ?
1. Kemungkinan besar Gradient Boosting mengalami overfit (Gradient Boosting sangat mudah untuk overfit).
2. Gradient Boosting mengurangi nilai bias.
3. Algoritma boosting mencari error dari sebelumnya untuk meminimalisirnya.
