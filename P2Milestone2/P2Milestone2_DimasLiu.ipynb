# Preprocess data & Eksplorasi Data :

```python
IMG_HEIGHT= 150
IMG_WIDTH= 150
batch_size= 126
```

```python
train_data = tf.keras.preprocessing.image_dataset_from_directory(
    train_dir,
    shuffle=True,
    batch_size=batch_size,
    image_size=(IMG_HEIGHT, IMG_WIDTH)
)
```

```python
test_data = tf.keras.preprocessing.image_dataset_from_directory(
  test_dir,
  image_size=(IMG_HEIGHT, IMG_WIDTH),
  batch_size=batch_size)
```

# Feature Engineering & Data Augmentation :

### Rotasi di sini di gunakan untuk augmentasi data, untuk memperkenalkan si model dengan pasti perbedaan antar setiap klasifikasi gambar.

### Rescale di sini berfungsi seperti MinMaxScaler, 255 merupakan nilai dari RGB (0 -> 255).

```python
train_datagen = ImageDataGenerator(rescale = 1. / 255,
                                  rotation_range= 40,
                                  width_shift_range = 0.2,
                                  height_shift_range = 0.2,
                                  shear_range= 0.2,
                                  zoom_range= 0.2,
                                  horizontal_flip= True,
                                  fill_mode = 'nearest'
)
train_data = train_datagen.flow_from_directory(train_dir,
                                               target_size=(150,150),
                                               class_mode='categorical',
                                               batch_size=126,
)
```

```python
validation_datagen = ImageDataGenerator(rescale = 1. / 255)
val_data = validation_datagen.flow_from_directory(val_dir,
                                                  target_size=(150,150),
                                                  class_mode='categorical',
                                                  batch_size=126,
)
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
test_data = test_datagen.flow_from_directory(test_dir,
                                             target_size=(150,150),
                                             class_mode="categorical",
                                             batch_size=126,
)
```

# Tujuan dari modelling :
- Meng-identifikasi dan meng-klasifikasi gambar dengan benar sesuai dari class (gunting, batu, kertas).

# Modelling :

### Penjelasan params & hyperparams :
- input_shape : tinggi, lebar dan dimensi gambar yang mau di input ke model.
### - Convolution 2D :
- activation : di sini akan di gunakan ReLU, supaya komputasi yang cepat (karena simple), dan tidak adanya terjadi masalah vanishing gradient di sini. Maka, ReLU akan aman dan bagus digunakan untuk model tersebut.
- kernel_size : ukuran kernel matrix untuk perkalian konvolusi yang diubah menjadi 2D.
### - MaxPooling 2D :
padding : valid memiliki asumsi kalau input gambar akan di-cover oleh filter, maka akan digunakan 'same' (pertimbangan dari image format PNG yang berarti background gambar kosong).
- cover yang dimaksudkan di sini adalah perkalian konvolusi matrix dari kernel.
### - Flatten :
- flatten : mengubah matrix menjadi 1D untuk output si model menjadi hasil prediksi.
### - Dropout :
dropout : disini, dropout digunakan untuk mencegah overfit dari model.
### - Output layer :
- Dense : 3, jumlah klasifikasi dari gambar (gunting, batu, kertas).
- Activation (sigmoid) : meskipun dataset merupakan multiclass problems, akan digunakan sigmoid, karena problem tersebut adalah image classification.
#### Jumlah unit dari neuron (input & hidden) dan jumlah hidden layer merupakan pertimbangan dari kompleksitas data.

```python
def create_model():
    model= Sequential()
    model.add(Conv2D(64,
        activation="relu",
        kernel_size=3,
        input_shape=(150,150,3)))
    model.add(MaxPooling2D(
        pool_size=(2,2),
        padding='same'))
    model.add(Conv2D(64,
        activation="relu",
        kernel_size=3))
    model.add(MaxPooling2D(
        pool_size=(2,2),
        padding='same'))
    model.add(Conv2D(128,
        activation='relu',
        kernel_size=3))
    model.add(MaxPooling2D(
        pool_size=(2,2),
        padding='same'))
    model.add(Conv2D(128,
        activation="relu",
        kernel_size=3))
    model.add(MaxPooling2D(
        pool_size=(2,2),
        padding='same'))
    model.add(Flatten())
    model.add(Dropout(0.5))
    model.add(Dense(512,
        activation='relu'))
    model.add(Dense(3, activation='sigmoid'))
    
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=3e-4), loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
model = create_model()
```

```python
model.summary()
```

```python
from keras.utils.vis_utils import plot_model
plot_model(model, show_shapes=True, show_layer_names=True, to_file='model_layer.png')
```

### Di sini, akan di gunakan early stopping, untuk mencegah overfitting dan training yang lebih cepat & efisien.

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
history = model.fit(train_data, validation_data=val_data, epochs=25, batch_size=batch_size, 
                    validation_steps=3, callbacks=[early_stopping])
```

```python
score = model.evaluate(test_data)
```

```python
plt.title('Accuracy History')
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend();
```

```python
plt.title('Loss History')
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Validation')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend();
```

# Model Inference :
Akan dilakukan beberapa inference dari :
- Warna kulit (bukan rasis, namun warna kulit bisa ber-efek ke RGB dan pixel tersebut untuk melihat apakah model betulan meng-identifikasi gambar dan klasifikasi dengan benar).
- Untuk batu, ada beberapa gaya berbeda (seperti kepalan tangan yang tidak tertutup dengan rapat), akan dilakukan 2 prediksi inference.
- Karena gunting memiliki 2 bentuk yang berbeda (dari segi ekspresi tangan), akan dilakukan 2 prediksi inference.

```python
def pred_infer(image, img_height, img_width):
    img = keras.preprocessing.image.load_img(image, target_size=(img_height, img_width))
    img_array = keras.preprocessing.image.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0)
    
    predict = model.predict(img_array)
    score = tf.nn.sigmoid(predict[0])
    print(f"Gambar ini merupakan {class_names[np.argmax(score)]} dengan persentase {np.max(score) * 100} %")
    return img
```
