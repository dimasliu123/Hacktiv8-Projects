```python
X = df.iloc[:, [0,1,2,3,4,5,6,9,10,12]]
y = df.iloc[:, 13]
```

# Library untuk model :

```python
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn import tree
from sklearn.ensemble import BaggingClassifier
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X,y, test_size= 0.2, random_state = 0)
```

# Decision Tree :

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc =  DecisionTreeClassifier(criterion='gini', class_weight='balanced')

params_dtc = {
    'max_depth' : [12, 13],
    'random_state' : [85, 95],
    'splitter' : ['best', 'random'],
    'min_samples_leaf' : [4, 5]
}

CV_dtc = GridSearchCV(dtc, params_dtc, scoring='recall', cv=5)

CV_dtc.fit(xtrain, ytrain)

dtc_best = CV_dtc.best_estimator_
dtc_pred = CV_dtc.predict(xtest) #model inference
```

```python
print("Best parameter : ", dtc_best)
print(classification_report(ytest, dtc_pred))
print(confusion_matrix(ytest, dtc_pred))
```

```python
plt.figure(figsize=(15,8))
tree = tree.plot_tree(dtc_best, filled=True)
```

# Random Forest :

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc =  RandomForestClassifier(criterion= 'gini', random_state = 50)

params_rfc = {
    'n_estimators' : [35, 45, 50],
    'max_depth' : [5,6,7],
    'class_weight' : ['balanced', 'balanced_subsample']
}
CV_rfc = GridSearchCV(rfc, params_rfc, scoring='recall', cv=5)

CV_rfc.fit(xtrain, ytrain)

rfc_best = CV_rfc.best_estimator_
rfc_pred = CV_rfc.predict(xtest) #model inference
```

```python
print("Best parameter : ", rfc_best)
print(classification_report(ytest, rfc_pred))
print(confusion_matrix(ytest, rfc_pred))
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
for i in range(len(CV_rfc.best_estimator_)):
    tree.plot_tree(CV_rfc.best_estimator_[i], filled=True)

plt.show()
```

```python
from sklearn.metrics import plot_roc_curve
```

```python
plot_roc_curve(rfc_best, xtest, ytest), plot_roc_curve(dtc_best, xtest,ytest)
```

```python
from sklearn.metrics import roc_auc_score
print("Random Forest ROC-AUC Score : ", roc_auc_score(ytest, rfc_pred))
print("Decision Tree ROC-AUC Score : ", roc_auc_score(ytest, dtc_pred))
```

```python
from sklearn.metrics import recall_score
print("Random Forest Recall Score : ", recall_score(ytest, rfc_pred))
print("Decision Tree Recall Score : ", recall_score(ytest, dtc_pred))
```

# Cross Validation Score :

```python
print("Decision Tree Cross Validation Recall Score (Train) : ", cross_val_score(dtc_best, xtrain, ytrain, scoring='recall').mean())
print("Decision Tree Cross Validation Recall Score (Test) : ", cross_val_score(dtc_best, xtest, ytest, scoring='recall').mean())
```

```python
print("Random Forest Cross Validation Score Train (Recall) : ",cross_val_score(rfc_best, xtrain, ytrain, scoring='recall', cv=5).mean())
print("Random Forest Cross Validation Score Test (Recall) : ",cross_val_score(rfc_best, xtest, ytest, scoring='recall', cv=5).mean())
```
